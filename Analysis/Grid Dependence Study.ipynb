```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_raw_1 = pd.DataFrame(pd.read_csv('data_unmodified_0.3x0.3_0.001.txt'))
data_raw_1.sort_values(by=['theta'], inplace=True)
data_1 = pd.DataFrame.to_numpy(data_raw_1).tolist()

data_10_df = pd.DataFrame(pd.read_csv('data_unmodified_0.2x0.2_0.001.txt'))
data_10_df = data_10_df.drop(['j'], axis = 1)
data_10_df.sort_values(by=['theta'], inplace=True)
data_10_df = data_10_df.dropna()
data_10 = pd.DataFrame.to_numpy(data_10_df).tolist()

#data_15_df = pd.DataFrame(pd.read_csv('data_unmodified_0.15x0.15_0.001.txt'))
data_15_df = pd.DataFrame(pd.read_csv('data_unmodifield_0.15x0.15_0.001_5points.txt'))
data_15_df = data_15_df.drop(['j'], axis = 1)
data_15_df.sort_values(by=['theta'], inplace=True)
data_15_df = data_15_df.dropna()
data_15 = pd.DataFrame.to_numpy(data_15_df).tolist()

exp_data = pd.read_csv('thesis data')
data_15
```
