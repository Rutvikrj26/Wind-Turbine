```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_15_df = pd.DataFrame(pd.read_csv('data_unmodified_0.0015_5x5.txt'))
data_20_df = pd.DataFrame(pd.read_csv('data_unmodified_0.002_5x5.txt'))
data_10_df = pd.DataFrame(pd.read_csv('data_unmodified_0.001_5x5.txt'))

data_15_df = data_15_df.drop(['j'], axis = 1)
data_10_df = data_10_df.drop(['j'], axis = 1)
data_20_df = data_20_df.drop(['j'], axis = 1)

data_15_df.sort_values(by=['theta'], inplace=True)
data_10_df.sort_values(by=['theta'], inplace=True)
data_20_df.sort_values(by=['theta'], inplace=True)

data_15_df = data_15_df.dropna()
data_10_df = data_10_df.dropna()
data_20_df = data_20_df.dropna()

data_15 = pd.DataFrame.to_numpy(data_15_df).tolist()
data_10 = pd.DataFrame.to_numpy(data_10_df).tolist()
data_20 = pd.DataFrame.to_numpy(data_20_df).tolist()
```

```python
theta_force_15 = {}
theta_force_cur_15 = {}

theta_force_10 = {}
theta_force_cur_10 = {}

theta_force_20 = {}
theta_force_cur_20 = {}

for i in range(len(data_15)):
    value = int((data_15[i][0]))
    value_limit = int((data_15[i][0])%360)
    theta_force_15[value] = data_15[i][4]#[7]
    theta_force_cur_15[value_limit] = data_15[i][4]
    
for i in range(len(data_10)):
    value = int((data_10[i][0]))
    value_limit = int((data_10[i][0])%360)
    theta_force_10[value] = data_10[i][4]#[7]
    theta_force_cur_10[value_limit] = data_10[i][4]
    
for i in range(len(data_20)):
    value = int((data_20[i][0]))
    value_limit = int((data_20[i][0])%360)
    theta_force_20[value] = data_20[i][4]#[7]
    theta_force_cur_20[value_limit] = data_20[i][4]    
```

```python
lists_15 = sorted(theta_force_15.items()) # sorted by key, return a list of tuples
lists_10 = sorted(theta_force_10.items()) # sorted by key, return a list of tuples
lists_20 = sorted(theta_force_20.items()) # sorted by key, return a list of tuples

x_15, y_15 = zip(*lists_15) # unpack a list of pairs into two tuples
x_10, y_10 = zip(*lists_10) # unpack a list of pairs into two tuples
x_20, y_20 = zip(*lists_20) # unpack a list of pairs into two tuples

plt.plot(np.array(x_15[1319:1679]), np.array(y_15[1319:1679])/(25.725*25) ,'--', color='blue', label = "timestep = 0.0015 sec")
plt.plot(np.array(x_10[1320:1680]), np.array(y_10[1320:1680])/(25.725*25), '--', color='orange',label = "timestep = 0.0010 sec")
#plt.plot(np.array(x_20[1291:1651]), np.array(y_20[1291:1651])/(25.725*25), '--', color='red' ,label = "timestep = 0.0020 sec")
                        
#plt.scatter(exp_data['theta'], exp_data['value'], label = "maclaren data")
plt.legend(framealpha=1, frameon=False, fontsize=12)
plt.xlabel(r'$\theta$', fontsize=20)
plt.ylabel('Thrust Coefficient, ' + r'$C_T$', fontsize=20)
plt.grid()
#plt.figtext(.3,.95,'Time', fontsize=30, ha='center')
#plt.figtext(.62,.95,'Independence',fontsize=25,ha='center')
```
